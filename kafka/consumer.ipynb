```python
# consumer_predictor_xgb.py
from kafka import KafkaConsumer
import json, joblib
import numpy as np
from collections import deque
```

```python
# 1. Load saved model
model = joblib.load("xgb_multi_model.pkl")
```

```python
# 2. Features (must match training order!)
FEATURES = ["temperature_2m (Â°C)", "relative_humidity_2m (%)", "wind_speed_10m (km/h)", "soil_temperature_0_to_7cm (Â°C)"]  

look_back, horizon = 72, 6
buffer = deque(maxlen=look_back)
```

```python
# 3. Kafka consumer
consumer = KafkaConsumer(
    "sensor-stream",
    bootstrap_servers="localhost:9092",
    value_deserializer=lambda m: json.loads(m.decode("utf-8")),
    auto_offset_reset="earliest",
    enable_auto_commit=True,
    group_id="ml-consumer-group"
)

print("âœ… Consumer started. Waiting for messages...")

for record in consumer:
    msg = record.value
    row = [msg.get(f, 0) for f in FEATURES]
    buffer.append(row)

    if len(buffer) == look_back:
        X_input = np.array(buffer).flatten().reshape(1, -1)
        y_pred = model.predict(X_input)
        y_pred = y_pred.reshape(horizon, len(FEATURES))

        out = {
            "current_time": msg.get("synthetic_timestamp"),
            "forecast_steps": [
                {f: float(y_pred[t, i]) for i, f in enumerate(FEATURES)}
                for t in range(horizon)
            ]
        }
        print("ðŸ”® Forecast:", json.dumps(out, indent=2))
```
