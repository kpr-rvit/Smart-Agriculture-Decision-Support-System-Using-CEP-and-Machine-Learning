```python
import pandas as pd
import numpy as np

def add_relative_noise(df, cols=None, std_percent=0.02, seed=42):
    np.random.seed(seed)
    if cols is None:
        cols = df.select_dtypes(include=[np.number]).columns.tolist()
    for c in cols:
        scale = df[c].std() * std_percent
        df[c] = df[c] + np.random.normal(0, scale, size=len(df))
    return df
```

```python
df = pd.read_csv("../datasets/latestdataset - Copy.csv")
```

```python
df.head()
```

```python
df.drop(columns=["rain (mm)", "precipitation (mm)", "soil_moisture_0_to_7cm (m³/m³)"], inplace=True)
```

```python
df.head()
```

```python
# Parse datetime
df['time'] = pd.to_datetime(df['time'])
df = df.set_index('time')
```

```python
# Convert all to float for scaling
df = df.astype(float)
```

```python
df.head()
```

```python
df = add_relative_noise(df, std_percent=0.01)  # change noise level here
df.to_csv("stream_ready.csv", index=False)
print("Wrote stream_ready.csv")
```
